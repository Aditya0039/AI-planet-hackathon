```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("hackathon_traindata_2.csv") 
print(df.shape)
```

```python
df.head()
#RIDAGEYR: Age in years i.e., Respondent's Age Group

#RIAGENDR: Respondent's Gender (1=Male, 2=Female)
#PAQ605: Physical activity questionnaire response: If the respondent engages in moderate or vigorous-intensity sports, fitness, or recreational activities in the typical week

#BMXBMI: Body Mass Index

#LBXGLU: Glucose level

#DIQ010: Diabetes questionnaire response

#LBXGLT: Glucose tolerance (Oral)

#LBXIN: Insulin level
```

```python
null_counts = df.isnull().sum()
print(null_counts)
```

```python
df.drop('SEQN', axis=1, inplace=True)
```

```python
df.head()
```

```python
df_cleaned = df[df.isnull().sum(axis=1) < 2]
```

```python
print(df.isnull().sum())
```

```python
df["RIAGENDR"]=df["RIAGENDR"].fillna(2)
```

```python
df["age_group"] = df["age_group"].fillna(df["RIAGENDR"].apply(lambda x: "Adult" if x < 65 else "Senior"))
```

```python
cols_to_fill = ["BMXBMI","LBXGLU","LBXGLT","LBXIN"]
for col in cols_to_fill:
    df[col] = df[col].fillna(
        df.groupby(['age_group','RIAGENDR'])[col].transform('mean')
    )
```

```python
df["PAQ605"]=df["PAQ605"].fillna(df["age_group"].apply(lambda x: 2 if x=="Adult" else 1))
```

```python
df["DIQ010"]=df["DIQ010"].fillna(df["age_group"].apply(lambda x: 1 if x=="Adult" else 2))
```

```python
print(df.isnull().sum())
df
```

```python
df.iloc[0:13,:]
```

```python
df["age_group1"]=df["age_group"].apply(lambda x: 0 if x=="Adult" else 1)
```

```python
df.drop('age_group', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
df
```

```python
df.rename(columns={'age_group1': 'age_group'}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.25,random_state=11)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
parameter={"penalty" : ["l1","l2","elasticnet"],"C":[1,2,3,4,10,20,30,40,50],"max_iter":[100,200,300]}
classifier_regressor=GridSearchCV(model,param_grid=parameter,scoring="accuracy",cv=5)
classifier_regressor.fit(X_train,y_train)
```

```python
print(classifier_regressor.best_params_)
```

```python
print(classifier_regressor.best_score_)
```
